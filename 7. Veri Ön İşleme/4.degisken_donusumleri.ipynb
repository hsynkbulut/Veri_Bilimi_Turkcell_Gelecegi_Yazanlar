### Değişken Standardizasyonu (Veri Standardizasyonu)

```python
import numpy as np
import pandas as pd
V1= np.array([1,3,6,5,7])
V2= np.array([7,7,5,8,12])
V3= np.array([6,12,5,6,14])
df = pd.DataFrame(
        {"V1" : V1,
        "V2" : V2,
        "V3" : V3})

df = df.astype(float)
df
```

### Değişken dönüşümüyle Değişken standardizasyonu arasındaki farkı bilmek çok önemlidir. Nedir ikisi arasındaki fark?
İkisi de değişkenin üzerinde bir dönüştürme işlemini ifade ediyor. 

---

#### Değişken standardizasyonu ne demektir?
Şöyle ki değişken standardizasyonundan bahsedildiğinde değişkenin kendi içindeki varyans yapısı, bilgi yapısı bozulmaz. Fakat belirli bir standarta oturtulur. Yani bu şu anlama geliyor örneğin değişken olduğu şekliyle durduğunda bir 10 değerini düşünelim. Bu 10 değeri örneğin veri seti küçükten büyüğe sıralandığında 80.sırada. Bu değişken standartlaştırıldığında bu 10 değeri muhtemelen 1,2 gibi bir değer olacak veya 0-1 arasında bir değer olacak. Fakat veri seti yine küçükten büyüğe sıralandığında bu değer yine 80.değere karşılık geliyor olacak. Dolayısıyla bir değişken standartlaştırıldığında değişkenin değerleri değişecektir, belirli bir formata sokulacaktır. Fakat taşımış olduğu yayılım, dağılım bilgisinin özütü (direk kendisi değil) mevcutta kalıyor olacaktır. Yani değişmeyecektir. Bu standartlaştırma işlemidir.

#### Peki değişken dönüştürmek ne demektir?
Değişken dönüştürmek ise dönüştürülen bir değişkenin taşıdığı bilginin taşındığı şekliyle kalamaması, dönüştürülmesi demektir.

---

#### Özetle
**Standartlaştırmak** veri setinin taşımış olduğu bilgiyi bozmadan belirli bir standarta getirir. Ama bilgiyi değiştirir, özütünü bozmadan belirli bir standarta getirir. 

**Dönüştürme işlemi** ise veri setinin taşımış olduğu bilgiyi bazen bozar bazen de temsil şeklini değiştirir.

#### Standartlaştırma İşlemleri
1. Standardizasyon
2. Normalizasyon
3. Min-Max Dönüşümü


### 1. Standardizasyon

```python
from sklearn import preprocessing

# scale(): dönüştür demek

# Bu işlem veri setinin içerisindeki bütün değişkenleri standartlaştırmış oldu
preprocessing.scale(df)
```

```python
df.mean()
```

### 2. Normalizasyon
Değerleri 0 ile 1 arasına dönüştürmek için kullanılır

```python
preprocessing.normalize(df)
```

### 3. Min-Max Dönüşümü
İstemiş olduğumuz iki tane aralığın arasına değerleri dönüştürmek istersek bu durumda Min-Max Dönüşümünü uygularız.

```python
scaler = preprocessing.MinMaxScaler(feature_range= (10, 20))
scaler.fit_transform(df)
```

### Çok Dikkat!
Eğer elinizdeki kategorik değişkenin 2'den fazla sınıfı varsa ve fonksiyonların bizden beklentisi doğrultusunda 
bu kategorik değişkeni bu şekilde nümerik bir formata çevirmek istiyorsak ve neticesinde böyle bir çevirme işlemi 
yaptıysak burası çok hassas bir nokta. Çünkü eğer elimizdeki bir değişkeni bu şekilde bir nümerik değerlere dönüştürür 
ve algoritmalara gönderirsek bu artık bunu 0 ile 3 arasındaki değerlerden oluşan bir değişken olarak tanımlayacak. 
Yani aslında bakacak olursak bu dönüşüm, algoritmaların kafasını karıştıracak ve kategorik değişkenin sınıflarının bağımlı 
değişkene olan etkileri bozulacaktır. Neden bozulmaktadır? Çünkü burada sınıflar nümerik değerlere dönüştü. Sanki aralarında
fark varmışcasına bir sıralama işlemine tutulmuş oldu. Normalde sınıfların arası eşit iken bu dönüşüm ile bu bozuldu ve 
örneğin 0'dan 10'a doğru giderken sanki bu değer artıyormuş gibi gözüküyor. 10 daha büyükmüş gibi gözüküyor. Halbuki bu 
nominal ölçek türü ile ölçülmüş bir kategorik değişkendi. İşte bu durum birçok probleme, yanlılıklara, hatalı çıkarımlara, 
gürültüye sebep olabilmektedir. Dolayısıyla algoritmaları aldatmakta, onları şaşırtmaktadır. 

İşte bu durumda One Hot Encoding yani One Hot Dönüşümü yapmış olacağız.

---

### One Hot Dönüşümü Ve Dummy Değişken Tuzağı

```python
df.head()
```

```python
# prefix: Yapılacak/oluşturulacak olan değişkenlerin ön isimlendirmesinin ne olacağını ifade etmek demek.
df_one_hot = pd.get_dummies(df, columns= ["sex"], prefix= ["sex"])
df_one_hot.head()
```

#### **Dummy Değişken Tuzağı Nedir?**
Eğer veri seti içerisindeki değişkenlere bir dönüşüm uyguladığımızda, eğer bu dönüşüm sonrasında oluşturulan yeni değişkenler birbirleri üzerinden oluşturulabiliyorsa bu durum Dummy Değişken Tuzağı olarak adlandırılır. Bir başka ifadeyle bir değişkeni ifade eden başka bir değişken varsa bu duruma Dummy Değişken Tuzağı denir.

Şimdi bakın yukarıdaki çıktıda sex_Male üzerinden sex_Female değişkeni oluşturulabilmektedir. Ya da tam tersi oluşturulabilmektedir. Yani aslında bakarsanız bu iki değişkenin ikisi de aynı bilgiyi taşımaktadır. Hem orijinal hali hem One Hot Dönüştürülmüş hali olarak üçü de aynı şeyi ifade ediyor olabilir. İşte bu duruma Dummy Değişken Tuzağı denir. 

##### **Peki bu durumu nasıl kontrol altında tutabiliriz?**
Şöyle ki oluşturulan değişken sayısından bir alt sayıda daha doğru ifadesi kategorik değişkenin sınıf sayısından daha az sayıda Dummy Değişken olması gerekiyor. Örneğin buradaki cinsiyetin 2 tane sınıfı vardı. Bu durumda 1 tane Dummy Değişken olmalı. Değişkenin orijinal hali veri setinde olmamalı.
