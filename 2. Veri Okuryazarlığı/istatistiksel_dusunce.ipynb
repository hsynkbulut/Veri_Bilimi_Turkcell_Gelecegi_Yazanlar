<a id = "3"></a><h1 id="Introduction"><span class="label label-default" style="background-color:#2E86C1; border-radius:8px; font-weight: bold; font-family:Verdana; font-size:22px; color:#FBFAFC; ">Kişiye Özgülük (Seviye 1) </span></h1>

*Genellikle bu basamakta kişiler veri setlerini şöyle bir bakıp verideki değişkenleri vb. bazı o ilk temas noktalarındaki yorumlamaları yapamıyor oluyor. Yani veriye bakıyor, ortalama var ama ne anlama geldiği bilinemeyebiliyor, standart sapmanın ne anlama geldiği bilinemeyebiliyor. Bu ilk ilkel seviye, başlangıç seviyesi.*

<a id = "4"></a><h1 id="Introduction"><span class="label label-default" style="background-color:#2E86C1; border-radius:8px; font-weight: bold; font-family:Verdana; font-size:22px; color:#FBFAFC; ">Geçici (Seviye 2) </span></h1>

*Bu nicel düşünmenin öneminin fark edilmeye başlandığı basamak ve hatta her zaman doğru olmasa da merkezi eğilim ve yayılım ölçüleri yorumlanabilir ve sayılarla ifade edilebilir bir hale geliyor. Ama veri üzerindeki bakış açısı tek yönlü. Veri temsilleri veya verinin analizine yönelik olarak bağlantıların çok kurulamadığı bir bölüm, seviye.*

<a id = "5"></a><h1 id="Introduction"><span class="label label-default" style="background-color:#2E86C1; border-radius:8px; font-weight: bold; font-family:Verdana; font-size:22px; color:#FBFAFC; ">Nicel (Seviye 3) </span></h1>

*Merkezi eğilim ve yayılım ölçülerinin doğru bir şekilde anlaşılmaya başlandığı basamak ve istatistiksel kararlar alabilmek için genellikle bu nicel verilere başvurabilme muhakemesinin pekiştiği basamak. Bağlam ve verinin ikisinin de bilincinde olunması ve bu kavramlar arasında ufak ufak ilişkiler kurulmasını ifade ediyor.*

<a id = "6"></a><h1 id="Introduction"><span class="label label-default" style="background-color:#2E86C1; border-radius:8px; font-weight: bold; font-family:Verdana; font-size:22px; color:#FBFAFC; ">Analitik (Seviye 4) </span></h1>

*Artık veriyi inceleme, analiz etme ve yorumlamada tam anlamıyla bir analitik yaklaşım kullanılıyor. Ortalama, medyan nedir, bunların birbiriyle farkları nedir, standart sapma, varyans nedir. Bunların gerçek hayatta bir ilişki kurduğumuzda neye karşılık geldiği bilinebilir bir hale geliyor. Dolayısıyla veri ve bağlam arasında ilişki kurulabiliyor ve bu ilişkiye kanıtlar gösterilebiliyor.*

<a id = "7"></a><h1 id="Introduction"><span class="label label-default" style="background-color:#2E86C1; border-radius:8px; font-weight: bold; font-family:Verdana; font-size:22px; color:#FBFAFC; ">Verinin Tanımlanması </span></h1>

***Örn:*** *Bir şirket internet kullanımı ile ilgili yaptığı araştırmada şu açıklamaları yayınlamıştır:*

*Anketi yanıtlayan 2000 kişinin %43,4'ünü erkekler, %66,4'sini kadınlar oluşturmaktadır. Anketi yanıtlayanların %80'i 15-27 yaş aralığındadır. Kadınların %72'si İngilizce bilmektedir. Erkekler günde ortalama 3 saat, kadınlar ise 4 saat internette zaman geçirmektedir. Erkeklerin %72'si, kadınların ise %75'i üniversite mezunudur.*

> ***A. Çalışmada ölçülmeye çalışılan değişkenler nelerdir?***
1. *Cinsiyet*
2. *Yaş*
3. *İngilizce Bilme*
4. *İnternette Geç. Zaman*
5. *Eğitim Durumu*

> ***B. Değişkenlerin türleri nelerdir?***
1. *Cinsiyet – Kategorik Değişken*
2. *Yaş – Sayısal Değişken*
3. *İngilizce Bilme - Kategorik Değişken*
4. *İnternette Geç. Zaman - Sayısal Değişken*
5. *Eğitim Durumu - Kategorik Değişken*

> ***C. Belirlediğiniz değişkenlerin hangi ölçekle ölçüldüğünü belirtiniz.***
1. *Cinsiyet – Kategorik Değişken - Nominal*
2. *Yaş – Sayısal Değişken - Oran*
3. *İngilizce Bilme - Kategorik Değişken - Nominal*
4. *İnternette Geç. Zaman - Sayısal Değişken - Oran*
5. *Eğitim Durumu - Kategorik Değişken – Ordinal*

> ***D. Metni ilk okuduğunuzda dikkatinizi çeken bir anormallik var mı?***

*Anketi yanıtlayanların kadın-erkek oranında anormallik var. Çünkü ikisi bir araya geldiğinde 100 oranını vermiyor. 2.anormal olan şey kadınların %72’si İngilizce bilmektedir ifadesidir. Buradaki anormallik metnin bize sunum tarzında bilgiyi tam açık olarak sunamama problemi vardır. Neden? Çünkü İngilizce’ye kıyasla başka bir dil var mı yoksa başka dillerin arasından İngilizce mi? Ya da İngilizce bilmek mi bilmemek mi? Gördüğünüz üzere bütün bu soruların yanıtı havada. İşte bu da bunun anormal olduğunu bize gösteriyor.*

> ***E. Bu veri seti üzerinde birkaç dakika yorum yapabilir misiniz?***

*Bunu kendinize göre yorumlayınız.*
