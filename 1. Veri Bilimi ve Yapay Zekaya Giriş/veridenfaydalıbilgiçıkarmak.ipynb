## 3- Tahminsel Analitik
Ne olacak? Sorusuna yanıt aranır. Nihai sonucu bilinmeyen olaylarla ilgili tahminler ya da gelecekle ilgili tahminler yapmak için kullanılır. Örneğin;



Ağustos ayında olduğumuzu ve bir sonraki 3 ayda neler olabileceğini tahmin etmek istediğimizi düşünelim. İşte bu durumda gelecekle ilgili henüz gerçekleşmemiş olan bir şeylere erişmeye çalışıyoruz. İşte bu durumlarda tahminsel analitik yaklaşımları kullanılır. Burada istatistiksel modellemeler, ekonometrik modellemeler ve makine öğrenmesi modellemeleri kullanılır.

## 4- Yönergeli Analitik 
Ne, nasıl olmalı? sorularına yanıt arar. Örneğin;



Gelirlerin önümüzdeki 3 ayda tahminleri bunlar ama ben bu gelirleri artırmak için ne yapmalıyım ne yaparsam bu gelirler artar? Sadece geleceği tahmin etmek yetmiyor dediğimizde Yönergeli Analitik yapmış oluyoruz. Aslında burada eğitecek olduğumuz makine öğrenmesi algoritmalarına ben ne yaparsam satışlarım artar sorusunu yöneltmiş oluyoruz ve bize vermiş olduğu yönergelere göre bazı iş aksiyon kararları alarak gelirlerimizi bu doğrultuda artırmaya çalışıyoruz.
