### *Line Plot*
Line Plot (Çizgi Grafiği): Zamana bağlı değişkenlerin olduğu veri setlerinde ve nesnelerin interneti gibi senaryolarda makinelerin ürettiği verileri görselleştirmek için kullanabileceğimiz bir veri görselleştirme tekniğidir.

#### "fmri" veri seti nedir?
Beyne bağlanan bir cihaz aracılığıyla toplanan sinyalleri ifade eden bir veri setidir.

- subject: verilerin toplandığı kişiler
- timepoint: zaman noktaları
- event: Birbirinden farklı olayları ifade ediyor
- region: Sinyalin toplandığı bölge
- signal: Gelen sinyali ifade ediyor

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count() 
# Buradaki çıktıdan anladığımız; her bir zaman noktasına göre eşit sayıda sayma işlemi yapılmış, 
# yani her bir zaman noktasında eşit sayıda sinyal toplanmış.
```

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe()
```

### *Creating a Line Plot*

```python
# Çizgi grafik ve çaprazlamalar
sns.lineplot(x= "timepoint", y= "signal", data= df);
```

```python
sns.lineplot(x= "timepoint", y= "signal", hue= "event", data= df);
```

```python
sns.lineplot(x= "timepoint", y= "signal", hue= "event", style= "event", data= df);
```

```python
# dashes argümanı, kesikli çizgi olup olmamasını sağlar grafikte. 

sns.lineplot(x = "timepoint", 
             y = "signal", 
             hue = "event", 
             style = "event", 
             markers= True, dashes= False, data= df);
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             hue = "region", 
             style = "event", 
             data= df);
```
