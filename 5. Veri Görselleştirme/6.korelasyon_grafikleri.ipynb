### *Correlation Matrix*
*Sürekli değişkenlerin arasındaki ilişkiyi görselleştirmek için kullanılır.*

*Yani örneğin elimizde iki değişken olduğunda bu değişkenlerin ilişkisi söz konusu olduğunda bu ilişki incelendiğinde buna Korelasyon adı verilir.*

*Scatter Plot bize sayısal değişkenler arasındaki ilişkiyi gösterir.*


- total_bill: yemeğin toplam fiyatı (bahşiş ve vergi dahil)
- tip: bahşiş
- sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)
- smoker: grupta sigara içen var mı? (0=No, 1=Yes)
- day: gün (3-Thur, 4-Fri, 5=Sat, 6=Sun)
- time: ne zaman? (0-Day, 1=Night)
- size: grupta kaç kişi var?

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
# scatterplot(): saçılım grafiğidir.
sns.scatterplot(x= "total_bill", y= "tip", data= df);
```

### *Kolerasyon Çaprazlamalar*

```python
sns.scatterplot(x= "total_bill", y= "tip", hue= "time", data= df);
```

```python
# style argümanı: style argümanına "time" değişkenini vererek grafiğin time değişkeninin 
# özelliklerine göre bölünme işlemi yapılması gerektiğini anlıyor. 
# Bölünmeden kasıt, simgelerin bölünmesi yani simgelerin farklı bir renklendirme ile yapılması gerektiğini anlıyor.
sns.scatterplot(x= "total_bill", y= "tip", hue= "time", style= "time", data= df);
```

```python
sns.scatterplot(x= "total_bill", y= "tip", hue= "day", style= "time", data= df);
```

```python
sns.scatterplot(x= "total_bill", y= "tip", hue= "size", size= "size", data= df);
```

### *Showing the Linear Relationship*

```python
import seaborn as sns
import matplotlib.pyplot as plt

# lmplot(): lineer model
# Doğrusal bir çizgi koyacağımız için lmplot() fonksiyonunu kullanacağız.
sns.lmplot(x= "total_bill", y= "tip", data= df);

# Grafikteki doğrusal çizginin etrafında olup çizgi boyunca uzayan açık renkli (saydam) şekil sapmaları ifade ediyor.
```

```python
sns.lmplot(x= "total_bill", y= "tip", hue= "smoker", data= df);
```

```python
sns.lmplot(x= "total_bill", y= "tip", hue= "smoker", col= "time", data= df);
```

```python
sns.lmplot(x= "total_bill", y= "tip", hue= "smoker", col= "time", row= "sex", data= df);
```

### *Scatter Plot Matrix*

*"iris" veri setinde bazı çiçek türleri var ve bu türlerin özelliklerini ifade eden değişkenler var. Yapraklarıyla ilgili, uzunluk, genişlik vb. şekilde bilgileri ifade eden bir veri seti.*

```python
import seaborn as sns
iris = sns.load_dataset("iris")
df = iris.copy()
df.head()
```

```python
df.dtypes
```

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue= "species");
```

```python
sns.pairplot(df, hue= "species", markers= ["o", "s", "D"]);
```

```python
# grafiğe doğru eklemek için kind= "reg" kullanalım
sns.pairplot(df, kind= "reg");
```

```python
sns.pairplot(df, kind= "reg", hue= "species");
```
